1. Escribe un programa que reciba un número e indique si es positivo.

```python
# Solicitar al usuario un número
numero=float(input("Ingrese un número: "))
# Indicar si el número es positivo, negativo o cero
if numero > 0:
    print("El número es positivo.")
elif numero < 0:
    print("El número es negativo.")
else:
    print("El número es cero.")
```

3. Escribe un programa que reciba dos números e imprima el mayor.

```python
# Solicitar dos números
numero1=float(input("Ingrese el primer número: "))
numero2=float(input("Ingrese el segundo número: "))
# Comparar los números e imprimir el mayor
if numero1 > numero2:
    print(f"El número mayor es: {numero1}")
elif numero2 > numero1:
    print(f"El número mayor es: {numero2}")
else:
    print("Ambos números son iguales.")
```

4. Escribe un programa que reciba una calificación (0-100) e indique si es Aprobado (>=60) o Reprobado.

```python
# Solicitar una calificación
calificacion=float(input("Ingrese la calificación de 0 a 100: "))
# Verificar si es aprobado o reprobado
if 0 <= calificacion <= 100:
    if calificacion >= 60:
        print("Aprobado")
    else:
        print("Reprobado")
else:
    print("Calificación inválida. Debe estar entre 0 y 100.")
```

9. Escribe un programa que lea un número e imprima “Fizz” si es divisible por 3, “Buzz” si es divisible por 5 y “FizzBuzz” si es divisible por ambos.

```python
numero=int(input("Escriba un número: "))
if numero % 3 == 0 and numero % 5 == 0:
    print("FizzBuzz")
elif numero % 3 == 0:
    print("Fizz")
elif numero % 5 == 0:
    print("Buzz")
else:
    print("El número no es divisible por 3 ni por 5.")
```
